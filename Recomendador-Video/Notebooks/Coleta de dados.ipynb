# Coleta de dados do youtube

Para a coleta de dados será utilizado a biblioteca **youtube_dl**, devido a facilidade do processamento dos dados e a posibilidade de eventuais correções na coleta para contornar mudanças de estrutura da página do Youtube.

Vamos importar as bibliotecas:

```python
#Importando as bibliotecas
import youtube_dl #biblioteca para pegar os dados do youtube
import pandas as pd
import numpy as np
```

Iremos utilizar o argumento *ignoreerrors* para contornar erros que ocorrem durante a coleta quando o vídeo se encontra indisponível.

```python
#Instanciando o objeto youtube_dl
ydl = youtube_dl.YoutubeDL({"ignoreerrors": True})
```

O próximo passo é definirmos os assuntos que iremos coletar os dados, rodar um loop e armazenar em uma lista.

```python
#queries com os assuntos a serem buscados pela biblioteca
queries = ["machine+learning", "data+science", "kaggle"]

#Criação de uma lista para armazenar os resultados da coleta
lista_info_videos = []
```

```python
#rodando o loop em relação aos assuntos
for query in queries:
    
    #Extrai as informações
    extract_info = ydl.extract_info("ytsearchdate2000:{query}".format(query = query), download = False) #ytsearchdate -> pega os vídeos mais recentes
    
    #Roda um for para cada entrada de dados -> entries (chave a onde possui as informações que precisamos)
    for entry in extract_info['entries'] :
        
        #Verifica se não é um vídeo indisponível (para estes casos retorna um valor vazio sem a chave entry)
        if entry is not None :
            
            #Armazena o assunto da query
            entry['query'] = query
    
    #Armazena as informações das buscas
    lista_info_videos += (extract_info['entries'])
```

```python
#Retira os valores vazios recebidos quando o vídeo está indisponível
lista_info_videos = [info for info in lista_info_videos if info is not None]
```

```python
#Transforma em um dataframe
df = pd.DataFrame(lista_info_videos)

#Imprime os 5 primeiros valores
df.head()
```

```python
#Convertendo string para datetime
df['upload_date'] = pd.to_datetime(df['upload_date'])
```

```python
#Definindo as colunas que serão utilizadas
colunas = ['title', 'upload_date', 'view_count', 'query']

#Extraindo os dados do tipo feather
df[colunas].to_feather("raw_data.feather")

#Extraindo as informações para o CSV
df[colunas].to_csv("raw_data_sem_label.csv")
```
